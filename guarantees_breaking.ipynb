```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import itertools
data = []
ids = set()
with open('data/guarantee_breaking_examples.txt') as f:
    for line1,line2 in itertools.zip_longest(*[f]*2):
        line = (line1 + line2).replace('\n', '').split('[')
        id = int(line[0])
        if id not in ids:
            data.append((id, eval(' '.join(('[' + line[1]).split()).replace(' ', ','))))
        ids.add(id)
```

```python
len(data)
```

```python
for i, (id, probs) in enumerate(data):
    plt.plot(np.arange(1, len(probs) + 1, 1), probs)
    if i % 20 == 0 and i > 0:
        plt.show()
```
